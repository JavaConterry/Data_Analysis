# Importing libs

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
```

## Clusterization using Hierarihical method

```python
df_scaled = pd.read_csv('c1.csv')
linkage_matrix = linkage(df_scaled.iloc[:, 1:], metric='euclidean')
```

### Dendogram visualization

```python
plt.figure(figsize=(12, 8))
dendrogram(linkage_matrix, 
           labels=df_scaled.index,
           leaf_font_size=10)



plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Index')
plt.ylabel('Distance')
plt.show()
```

## Define a "Class" of given input using KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv('c2.csv')

X, y = df.iloc[:,1:], df.iloc[:, 0]
clf = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
clf.fit(X, y)

clf.predict([[429,2,46,2021]])
```

### Result of predicting [429,2.46,2021] - class 1
