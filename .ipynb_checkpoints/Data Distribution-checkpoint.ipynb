## Bootstrapping

```python
import numpy as np
import pandas as pd
from scipy.stats import norm

data = df['Income']
n_bootstrap = 1000

def bootstrap(data, n_bootstrap):
    bootstrap_samples = [np.mean(np.random.choice(data, size=len(data), replace=True)) for _ in range(n_bootstrap)]
    return np.array(bootstrap_samples)

bootstrap_means = bootstrap(data, n_bootstrap)

alpha = 0.05 
lower = np.percentile(bootstrap_means, alpha / 2 * 100)
upper = np.percentile(bootstrap_means, (1 - alpha / 2) * 100)

print(f"Mean: {np.mean(data):.2f}")
print(f"95% Confidance Interval: [{lower:.2f}, {upper:.2f}]")

plt.figure(figsize=(10, 6))
sns.histplot(bootstrap_means, kde=True, color='blue')
plt.axvline(lower, color='red', linestyle='--', label=f'Lower CI: {lower:.2f}')
plt.axvline(upper, color='green', linestyle='--', label=f'Upper CI: {upper:.2f}')
plt.title("Bootstrap Distribution with 95% Confidence Interval")
plt.xlabel("Mean")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
